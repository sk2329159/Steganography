```python
import cv2
import os

# Define the image path
image_path = "tiger.jpg"  # Update with the correct path if needed

# Debugging: Check if the image exists
if not os.path.exists(image_path):
    print(f"Error: Image '{image_path}' not found in directory: {os.getcwd()}")
    print("Available files:", os.listdir())  # Show all files in the directory
    exit()

# Load the image
img = cv2.imread(image_path)

# Verify successful loading
if img is None:
    print("Error: Failed to read image. Ensure the image is not corrupted or an unsupported format.")
    exit()

# Make the image writable
img = img.copy()

# User input for message and passcode
msg = input("Enter secret message: ") + "#"  # Add delimiter to indicate end
password = input("Enter a passcode: ")

# Convert message into ASCII values
msg_length = len(msg)
height, width, _ = img.shape

# Check if the image can store the message
if msg_length > height * width:
    print("Error: Message too long for the image size.")
    exit()

# Encoding message into the image
idx = 0
for y in range(height):
    for x in range(width):
        if idx < msg_length:
            img[y, x, 0] = ord(msg[idx])  # Store ASCII in the Blue channel
            idx += 1
        else:
            break
    if idx >= msg_length:
        break

# Save the encoded image
cv2.imwrite("encryptedImage.jpg", img)
print("Message successfully encoded into 'encryptedImage.jpg'.")
```

```python
# Decryption process
message = ""
pas = input("Enter passcode for decryption: ")

if password == pas:
    idx = 0
    for y in range(height):
        for x in range(width):
            char = chr(img[y, x, 0])  # Read from Blue channel
            if char == "#":  # Stop at delimiter
                break
            message += char
            idx += 1
        if char == "#":
            break
    print("Decrypted message:", message)
else:
    print("Authentication failed. Access denied.")
```
